```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import pickle
import os
import sigfig
```

```python
file_name = 'C:\\Users\\John\\Documents\\Research\\ML_Gravity\\sh_stats_full_grid.data'

with open(file_name, 'rb') as f:
    df = pickle.load(f)

nn_data_name = 'C:\\Users\\John\\Documents\\Research\\ML_Gravity\\continuous_results_v2.data'

with open(nn_data_name, 'rb') as f:
    nn_df = pickle.load(f)
os.makedirs("C:\\Users\\John\\Documents\\Research\\ML_Gravity\\Plots\\World\\", exist_ok=True)
```

```python
scale=2
axes = df.plot(x=None, y=['sigma_2_mean', 'sigma_2_c_mean'],  subplots=True, layout=(1,2), figsize=(10*scale, 3.5*scale), grid=True)

data = nn_df.loc['Sun Nov 15 06:14:11 2020']

axes[0][0].axhline(y=data['sigma_2_mean'], label='non-PINN-new', c='g', linestyle='--') 
axes[0][1].axhline(y=data['sigma_2_c_mean'], label='non-PINN-new', c='g', linestyle='--')  

data = nn_df.loc['Sun Nov 15 07:34:43 2020']

axes[0][0].axhline(y=data['sigma_2_mean'], label='PINN-new', c='r', linestyle='--') 
axes[0][1].axhline(y=data['sigma_2_c_mean'], label='PINN-new', c='r', linestyle='--')
 
#data = nn_df.loc['Mon Nov  2 02:33:21 2020']
data = nn_df.loc['Thu Oct 29 22:11:15 2020']

axes[0][0].axhline(y=data['sigma_2_mean'], label='non-PINN-old', c='g') 
axes[0][1].axhline(y=data['sigma_2_c_mean'], label='non-PINN-old', c='g')  

#data = nn_df.loc['Mon Nov  2 03:57:02 2020']
data = nn_df.loc['Sat Oct 31 17:42:14 2020']


axes[0][0].axhline(y=data['sigma_2_mean'], label='PINN-old', c='r') 
axes[0][1].axhline(y=data['sigma_2_c_mean'], label='PINN-old', c='r')
 

plt.legend()
```
