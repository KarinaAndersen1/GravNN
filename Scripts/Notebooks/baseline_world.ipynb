```python
import pickle
time = pd.Timestamp('Sun Nov 15 06:14:11 2020').to_julian_date()
with open(os.path.abspath('.') +"\\..\\..\\Plots\\"+ str(time) +"\\loss.data", 'rb') as f:
    loss_adam_non_pinn = pickle.load(f)
    loss_lbgdfs_non_pinn = pickle.load(f)

time = pd.Timestamp('Sun Nov 15 07:34:43 2020').to_julian_date()
with open(os.path.abspath('.') +"\\..\\..\\Plots\\"+ str(time) +"\\loss.data", 'rb') as f:
    loss_adam_pinn = pickle.load(f)
    loss_lbgdfs_pinn = pickle.load(f)
```

```python
scale=2
plt.figure(figsize=(10.0*scale,3.5*scale))
plt.subplot(121)
epochs = np.linspace(1,len(loss_adam_non_pinn),len(loss_adam_non_pinn))*10
start = 20
plt.semilogy(epochs[start:], loss_adam_non_pinn[start:], label='non-PINN')
plt.semilogy(epochs[start:], loss_adam_pinn[start:], label='PINN')

plt.ylabel("Loss (Adam)")
plt.xlabel("Epochs")
plt.legend()

plt.subplot(122)
epochs = np.linspace(1,len(loss_lbgdfs_non_pinn),len(loss_lbgdfs_non_pinn))
pinn_epochs = np.linspace(1,len(loss_lbgdfs_pinn),len(loss_lbgdfs_pinn))

start = 2
plt.semilogy(epochs[start:], loss_lbgdfs_non_pinn[start:],label='non-PINN')
plt.semilogy(pinn_epochs[start:], loss_lbgdfs_pinn[start:], label='PINN')

plt.ylabel("Loss (L-BFGS-B)")
plt.xlabel("Epochs")
plt.legend()
plt.savefig("C:\\Users\\John\\Documents\\Research\\ML_Gravity\\Plots\\World\\optimizer_losses.pdf")
```
