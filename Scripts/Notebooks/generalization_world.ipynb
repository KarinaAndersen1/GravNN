```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import pickle
import sigfig
import os
```

```python


nn_data_name = 'C:\\Users\\John\\Documents\\Research\\ML_Gravity\\generalization_results.data'
with open(nn_data_name, 'rb') as f:
    nn_df = pickle.load(f)

os.makedirs("C:\\Users\\John\\Documents\\Research\\ML_Gravity\\Plots\\World\\", exist_ok=True)
```

```python
nn_df[nn_df['map_radius'] == 6378136.6+10000][nn_df['PINN_flag'] == True].iloc[0].name
```

```python
file_name = 'C:\\Users\\John\\Documents\\Research\\ML_Gravity\\sh_stats_world_R100.data'
with open(file_name, 'rb') as f:
    df = pickle.load(f)

scale = 2
axes = df.plot(x=None, y=['sigma_2_mean', 'sigma_2_c_mean'],  subplots=True, layout=(1,2), figsize=(5*scale, 3.5*scale), grid=True)

data = nn_df.loc['Sun Nov 15 20:05:16 2020']
axes[0][0].axhline(y=data['sigma_2_mean'], label='non-PINN',c='g') 
axes[0][1].axhline(y=data['sigma_2_c_mean'], label='non-PINN',c='g')  

data = nn_df.loc['Sun Nov 15 23:14:20 2020']
axes[0][0].axhline(y=data['sigma_2_mean'], label='PINN',c='r') 
axes[0][1].axhline(y=data['sigma_2_c_mean'], label='PINN',c='r')  
plt.legend()
plt.savefig("C:\\Users\\John\\Documents\\Research\\ML_Gravity\\Plots\\World\\generalization_error_100.pdf")
```
